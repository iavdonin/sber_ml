**35\.** Попробуйте реализовать метод Interquartile Range (IQR) (★★★)

Почитать про него можно по ссылке выше.

```python
s = pd.Series(np.logspace(-2, 2, 30))

#your code

q1 = s.quantile(q=0.25)
q3 = s.quantile(q=0.75)
iqr = q3 - q1
coef = 1.5

s[s < q1 - iqr * coef] = q1 - iqr * coef
s[s > q3 + iqr * coef] = q3 + iqr * coef

s
```

5. Для столбцов у которых значений более 10, оставляем топ 10 значений, остальные заменяем на 'other'

```python
n = 10

cat_cols = df.select_dtypes(exclude='number').columns

gt_10_unique = df[df[cat_cols].columns[df[cat_cols].nunique() > n]]

for col in gt_10_unique:
    top_10_values = df[col].value_counts().nlargest(n).index
    df[col][~df[col].isin(top_10_values)] = 'other'

df
```

6. Числовые столбцы проверяем на пропуски. Подсчитываем кол-во nan\none и делим на размер датафрейма -сохраняем в отдельный датафрейм, это доля пропусков в каждом столбце

```python
num_cols = df.select_dtypes(include='number').columns
nan_part_df = pd.DataFrame(df[num_cols].isna().sum() / len(df), columns=['nan_part'])
nan_part_df
```

7. Чистим выбросы. Во всех числовых столбцах заменяем все значения столбца меньше 5го персентиля и больше 95ого персентиля на эти пограничные значения.

```python
# Я не знаю, почему здесь вылетает warning, значения меняются корректно

num_cols = df.select_dtypes(include='number').columns
for col in num_cols:
    col_series = df[col]
    q05 = col_series.quantile(q=0.05)
    q95 = col_series.quantile(q=0.95)
    col_series.loc[col_series < q05] = q05
    col_series.loc[col_series > q95] = q95
    df.loc[:, col] = col_series
df
```

8. Заполняем пропуски в числовых столбцах на среднее по столбцу, в текстовых столбцах на 'other'

```python
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

cat_cols = df.select_dtypes(exclude='number').columns
df[cat_cols] = df[cat_cols].fillna('other')

df
```

9. Все категориальные признаки превращаем в one-hot-encoding
